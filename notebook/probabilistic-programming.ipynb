## Bayes theorem

$$
p(\theta | X) = \frac{p(\theta, X)}{p(X)} = \frac{p(X | \theta)p(\theta)}{p(X)}
$$

$$
\text{posterior} = \frac{\text{likelihood} \times \text{prior}}{\text{evidence}}
$$

## Intractable

$$
p(\theta | X) = \frac{p(X | \theta)p(\theta)}{p(X)} = \frac{p(X | \theta)p(\theta)}{\int p(X | \theta)p(\theta) d \theta}
$$

$$
\int p(X | \theta)p(\theta) d \theta
$$

## Approximation methods

* Markov Chain Monte Carlo (Sampling)
    * Large computation time
    * More exact
* Variational inference
    * Fast
    * Not exact

## Variational Inference

* Mean-field approximation
* Evidence Lower Bound Method (ELOB)

## Markov Chain Monte Carlo (MCMC)

* Rejection method
* Metropolis(-Hasting) method
* Gibbs sampling
* Slice method
* Hamiltonian Monte Carlo
* No-U-Turn Sampler (NUTS)

### Load data

```julia
data = RDatasets.dataset("datasets", "mtcars");
first(data, 6)
```

### Split our dataset 70%/30% into training/test sets.

```julia
train, test = MLDataUtils.splitobs(nrow(data), at = 0.7);
```

```julia
features = filter(x -> !in(x, [:MPG, :Model]), names(data))
train_X = Matrix(data[train, features])
test_X = Matrix(data[test, features])
train_y = data[train, :MPG]
test_y = data[test, :MPG]
```

```julia
standardize(x) = (x .- mean(x, dims=1)) ./ std(x, dims=1), x

unstandardize(x, orig) = x .* std(orig, dims=1) .+ mean(orig, dims=1)
```

```julia
# Standardize our dataset.
(train_X, train_orig) = standardize(train_X)
(test_X, test_orig) = standardize(test_X)
(train_y, train_l_orig) = standardize(train_y)
(test_y, test_l_orig) = standardize(test_y);
```

### Model

```julia
@model linear_regression(x, y, n_obs, n_vars) = begin
    # Set variance prior.
    σ₂ ~ TruncatedNormal(0, 100, 0, Inf)
    
    # Set intercept prior.
    intercept ~ Normal(0, 3)
    
    # Set the priors on our coefficients.
    coefficients = Array{Real}(undef, n_vars)
    coefficients ~ [Normal(0, 10)]
    
    # Calculate all the mu terms.
    mu = intercept .+ x * coefficients
    for i = 1:n_obs
        y[i] ~ Normal(mu[i], σ₂)
    end
end;
```

### Sampling

```julia
Turing.turnprogress(false);
```

```julia
n_obs, n_vars = size(train_X)
model = linear_regression(train_X, train_y, n_obs, n_vars)
chain = sample(model, NUTS(1500, 200, 0.65));
```

### Summary

```julia
describe(chain)
```

### Plot

```julia
plot(chain)
```

## Bayesian Logistic Regression

```julia
using StatsFuns: logistic
```

### Load data

```julia
data = RDatasets.dataset("ISLR", "Default");
first(data, 6)
```

### Preproccessing

```julia
# Create new rows, defualted to zero.
data[:DefaultNum] = 0.0
data[:StudentNum] = 0.0
```

```julia
isyes(x) = x == "Yes" ? 1.0 : 0.0

data[:DefaultNum] = isyes.(data.Default)
data[:StudentNum] = isyes.(data.Student);
```

```julia
# Delete the old columns which say "Yes" and "No".
deletecols!(data, [:Default, :Student])
first(data, 6)
```

### Training / testing set

```julia
train, test = MLDataUtils.splitobs(nrow(data), at=0.05);
```

```julia
label = :DefaultNum
features = [:StudentNum, :Balance, :Income]
train_X = Matrix(data[train, features])
test_X = Matrix(data[test, features])
train_y = data[train, label]
test_y = data[test, label];
```

```julia
train_X, _ = standardize(train_X)
test_X, _ = standardize(test_X)
```

### Model

```julia
@model logistic_regression(x, y, n, σ²) = begin
    intercept ~ Normal(0, σ²)

    student ~ Normal(0, σ²)
    balance ~ Normal(0, σ²)
    income  ~ Normal(0, σ²)

    for i = 1:n
        v = logistic(intercept + student*x[i, 1] + balance*x[i, 2] + income*x[i, 3])
        y[i] ~ Bernoulli(v)
    end
end;
```

### Sampling

```julia
# This is temporary while the reverse differentiation backend is being improved.
Turing.setadbackend(:forward_diff)
```

```julia
n, _ = size(train_X)  # Retrieve the number of observations
chain = sample(logistic_regression(train_X, train_y, n, 1), HMC(1500, 0.05, 10))
```

### Summary

```julia
describe(chain)
```

### Plot

```julia
# Extract the parameters we want to plot.
w1 = chain[:student]
w2 = chain[:balance]
w3 = chain[:income]

# Show the corner plot.
cornerplot(hcat(w1, w2, w3), compact=true, labels=[:student, :balance, :income])
```

## Gaussian Mixture Model

### Construct the data points

```julia
N = 30

μs = [-3.5, 0.0]

x = mapreduce(c -> rand(MvNormal([μs[c], μs[c]], 1.), N), hcat, 1:2)
```

### Visualize data

```julia
scatter(x[1,:], x[2,:], legend = false, title = "Synthetic Dataset")
```

### Model

$$
\mu_k \sim Normal(0, 1), \forall k \\
w \sim Dirichlet(k, \alpha) \\
\\
z_i \sim Categorical(w), \forall i \\
x_i \sim Normal(\mu_{zi}, 1.), \forall i
$$

```julia
@model GaussianMixtureModel(x) = begin
    D, N = size(x)

    μ1 ~ Normal()
    μ2 ~ Normal()
    μ = [μ1, μ2]
    
    α = 1.0
    w ~ Dirichlet(2, α)
    
    # Draw assignments for each datum and generate it from a multivariate normal.
    k = Vector{Int}(undef, N)
    for i in 1:N
        k[i] ~ Categorical(w)
        x[:,i] ~ MvNormal([μ[k[i]], μ[k[i]]], 1.)
    end
    return k
end
```

### Sampling

```julia
Turing.setadbackend(:forward_diff)
```

```julia
gmm_model = GaussianMixtureModel(x);
gmm_sampler = Gibbs(100, PG(100, 1, :k, :w), HMC(1, 0.05, 10, :μ1, :μ2))
tchain = sample(gmm_model, gmm_sampler);
```

### Visualize

```julia
ids = findall(map(name -> occursin("μ", name), tchain.names));
plot(Chains(tchain.value[:, ids, :], names = ["mu 1", "mu 2"]), colordim = :parameter, legend = true)
```

### Prediction distribution

```julia
function predict(x, y, w, μ)
    # Use log-sum-exp trick for numeric stability.
    return Turing.logaddexp(
        log(w[1]) + logpdf(MvNormal([μ[1], μ[1]], 1.), [x, y]), 
        log(w[2]) + logpdf(MvNormal([μ[2], μ[2]], 1.), [x, y])
    )
end
```

### Prediction distribution

```julia
contour(range(-5, stop = 3), range(-6, stop = 2), 
    (x, y) -> predict(x, y, [0.5, 0.5], [mean(tchain[:μ1]), mean(tchain[:μ2])])
)
scatter!(x[1,:], x[2,:], legend = false, title = "Synthetic Dataset")
```

### Infered Assignments

```julia
scatter(x[1,:], x[2,:], legend = false, title = "Assignments on Synthetic Dataset", zcolor = mean(tchain[:k]))
```
